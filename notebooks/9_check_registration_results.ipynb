```python
moving_path = 'round2/registered2/1_1_1'
z = 1500

moving_img = io.zarr.open(os.path.join(working_dir,
                                       moving_path),
                          mode='r')
moving_slice = moving_img[z]
moving_slice.shape
```

```python
fixed_path = 'round1/syto16.zarr/1_1_1'
z = 1500

fixed_img = io.zarr.open(os.path.join(working_dir,
                                       fixed_path),
                          mode='r')
fixed_slice = fixed_img[z]
fixed_slice.shape
```

```python
%matplotlib notebook
reg_slice = moving_slice

# Show an overlay of the registered and fixed slices
fixed_clim = [0, 4000]
reg_clim = [0, 8000]
figsize = (6, 6)

plotting.plot_overlay(fixed_slice, 
                      reg_slice, 
                      fixed_clim, 
                      reg_clim, 
                      figsize)
```
