```python
subimg = img[300:316].astype(np.float32)/4095
```

```python
img_frangi = frangi.frangi(subimg, 
                           scale_range=(1, 3),
                           alpha=0.5,
                           beta=0.5,
                           frangi_c=50,
                           black_vessels=False)
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img_frangi[2])
plt.show()
```

```python
g = seg.gaussian(subimg, sigma=(1.0, 2.0, 2.0))
```

```python
img_grad = seg.gradient(g, 2, 1, 1)
img_grad.shape
```

```python
S = seg.structure_tensor(img_grad)
S.shape
```

```python
eigvals = np.linalg.eigvalsh(S)
eigvals.shape
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(eigvals[4, ..., 2])
plt.show()
```

```python
t1 = eigvals[..., 2]
t2 = eigvals[..., 1]
t3 = eigvals[..., 0]
t1.shape
```

```python
N =  (t3-t1)**2
D = t1**2 + t2**2 + t3**2
fa = np.sqrt(0.5*N/D)
fa.shape
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(fa[4])
plt.show()
```
